```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com/catalogue/page-1.html"
```

```python
url
```

```python
response = requests.get(url) #to get current the web page
response
```

```python
#but we want the content of this web page
response_content = response.content
response_content
```

```python
#let's convert this content in html file because this is not the html
#BeautifulSoup is a Python library that is commonly used for web scraping and parsing HTML and XML documents.
#It provides a convenient way to extract data from web pages by navigating and searching through the HTML or XML structure of the page.
#BeautifulSoup makes it easier to work with web data and extract specific information, such as text, links, images, and other elements.
soup = BeautifulSoup(response_content, 'html.parser')
soup
```

```python
ol = soup.find('ol')
ol
```

```python
articles = ol.find_all('article', class_='product_pod')
articles
```

```python
for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    print(title)
```

```python
for article in articles:
    starTag = article.find('p')
    star = starTag['class'][1]
    print(star)
```

```python
for article in articles:
    price = article.find('p', class_='price_color').text
    price = float(price[1:])
    print(price)
```

```python
books = []
```

```python
for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    starTag = article.find('p')
    star = starTag['class'][1]
    price = article.find('p', class_='price_color').text
    price = float(price[1:])
    books.append([title, star, price])
books
```

```python
more_books = []

for i in range(1,5):
  url = f"https://books.toscrape.com/catalogue/page-{i}.html"
  response = requests.get(url)
  response = response.content
  soup = BeautifulSoup(response, 'html.parser')
  ol = soup.find('ol')
  articles = ol.find_all('article', class_='product_pod')
  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    starTag = article.find('p')
    star = starTag['class'][1]
    price = article.find('p', class_='price_color').text
    price = float(price[1:])
    more_books.append([title, star, price])
```

```python
more_books
```

```python
df = pd.DataFrame(more_books, columns=['Title', 'Star Rating', 'Price'])
df
```

```python
df.to_csv('books.csv')
```
